# Evaluation of pair trading

```python
import importlib
import S3_Pairs_trading

import pandas as pd

S3_Pairs_trading = importlib.reload(S3_Pairs_trading)
```

```python
def evaluate(String_array):
    rows = []
    for left, right in String_array:
        revenue_analysis, trades_analysis, revenue_test, trades_test = S3_Pairs_trading.main(left, right)
        rows.append((left, right, revenue_analysis, revenue_test))

        revenue_analysis, trades_analysis, revenue_test, trades_test = S3_Pairs_trading.main( right, left)
        rows.append((right, left, revenue_analysis, revenue_test))
    df = pd.DataFrame(
        rows,
        columns=["left", "right", "revenue analysis", "revenue test"]
    )
    return(df)
```

# Pairs trading suggested by Cointegration in the Dickey-Fuller test
Here we perform the pairs trading with the highest cointegration pairs

```python
results_coint=evaluate([['MU','MPWR'], ['MU','KLAC'],['QRVO','MCHP'],['TXN','ADI'],['TER','TXN']])
display(results_coint)

print('average revenue analysis: ',float(results_coint['revenue analysis'].mean()))
print('average revenue test for all: ',float(results_coint['revenue test'].mean()))
print('average revenue test with revenue in analysis: ',float(results_coint[results_coint['revenue analysis']>0]['revenue test'].mean()))
corr=results_coint.loc[:,['revenue analysis','revenue test']].corr()
print('Correlation analysis & test:',corr.loc['revenue analysis','revenue test'])
```
