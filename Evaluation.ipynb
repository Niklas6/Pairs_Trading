# Evaluation of pair trading

```python
import importlib
import Pairs_trading


import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

importlib.reload(Pairs_trading)
```

```python
def evaluate(String_array):

    rows = []

    for left, right in String_array:
        revenue_analysis, trades_analysis, revenue_trading, trades_trading = Pairs_trading.main(left, right)
        rows.append((left, right, revenue_analysis, revenue_trading))

        revenue_analysis, trades_analysis, revenue_trading, trades_trading = Pairs_trading.main( right, left)
        rows.append((right, left, revenue_analysis, revenue_trading))



    df = pd.DataFrame(
        rows,
        columns=["left", "right", "revenue analysis", "revenue test"]
    )

    return(df)




```

# Pairs Trading NVDA
Here we perform the pairs trading with NVIDA since it had the highest Correlation

```python
#[['NVDA','AVGO'],['NVDA','MPWR'],['LRCX','NVDA'],['KLAC','NVDA'],['MRVL','NVDA']]

results_NVDA=evaluate([['NVDA','AVGO'],['NVDA','MPWR'],['LRCX','NVDA'],['KLAC','NVDA'],['MRVL','NVDA']])
display(results_NVDA)

print('average revenue analysis: ',float(results_NVDA['revenue analysis'].sum())/10)
print('average revenue test: ',float(results_NVDA['revenue test'].sum())/10)




```
